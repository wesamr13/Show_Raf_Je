```python
df = r_cusine.join(df1, on = 'placeID', how = 'inner').set_index('placeID').join(u_ratings,'placeID','inner').drop(columns = ['latitude','longitude','the_geom_meter','fax','zip','url','other_services','address','country','accessibility','franchise'])
```

```python
def func(ser,val):
    ar = []
    x = ser.value_counts()
    for i in range(len(x)):
        if x.values[i] > val:
            ar.append(x.keys()[i])  
    return ar       
```

```python
df = df.loc[df['Rcuisine'].isin(func(df['Rcuisine']))]
```

```python
df_rate = df.groupby(['Rcuisine']).mean()
df_rate
```
